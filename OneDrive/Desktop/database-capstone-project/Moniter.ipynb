```python
import mysql.connector as connector
import time
```

```python
connection = connector.connect(
    user="root",
    password="Shivani@123",
    host="localhost",            
    database="little_lemon_db"
)
```

```python
TABLE_NAME = "bookings"
KEY_COLUMN = "BookingId"  

 
def process_all_existing_rows(cursor):
    print("Fetching all existing rows...")
    cursor.execute(f"SELECT * FROM {TABLE_NAME} ORDER BY {KEY_COLUMN}")
    rows = cursor.fetchall()
    print(f"Total rows fetched: {len(rows)}")
    if rows:
        for row in rows:
            print(row)
        return rows[-1][KEY_COLUMN]
    else:
        print("No existing rows found.")
        return 0

def monitor_table():
    cursor = connection.cursor(dictionary=True)
    last_max_id = process_all_existing_rows(cursor)
    print(f"Starting monitoring from BookingId > {last_max_id}")

    try:
        while True:
            cursor.execute(
                f"SELECT * FROM {TABLE_NAME} WHERE {KEY_COLUMN} > %s ORDER BY {KEY_COLUMN}",
                (last_max_id,)
            )
            new_rows = cursor.fetchall()
            print(f"New rows fetched: {len(new_rows)}")

            if new_rows:
                for row in new_rows:
                    print(row)
                last_max_id = new_rows[-1][KEY_COLUMN]
            else:
                print("No new rows found.")

            time.sleep(3)

    except KeyboardInterrupt:
        print("\nStopped monitoring.")
    finally:
        cursor.close()
        connection.close()

if __name__ == "__main__":
    monitor_table()


```
